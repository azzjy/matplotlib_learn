```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,100,100) #生成从0到10生成100个等列数差
plt.scatter(x,np.sin(x))
plt.show()
```

```python
'''
绘制10种大小100种颜色的散点图
'''
```

```python
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
color = np.random.rand(100)
size = np.random.rand(100)*1000
plt.scatter(x,y,s=size,c=color,alpha=0.7)
plt.show()
```

```python
#不同种类不同颜色的线 
x=np.linspace(0,10,100)
plt.plot(x,x+0,'-g') #实线 绿色
plt.plot(x,x+1,'--c')#虚线 浅蓝色
plt.plot(x,x+2,'-.k') #点划线 黑色
plt.plot(x,x+3,'-r')#实线 红色
plt.plot(x,x+4,'o') #点 默认是蓝色
plt.plot(x,x+5,'x') #叉叉 默认是蓝色
plt.plot(x,x+6,'d') #砖石 红色
plt.show()
   
```

```python
'''
添加图例
'''
#点 默认是蓝色
#叉叉 默认是蓝色 #砖石 红色
x=np.linspace(0,10,100)
plt.plot(x,x+0,'-g',label='-g')
plt.plot(x,x+1,'--c',label='--c')
plt.plot(x,x+2,'-.k',label='-.k')
plt.plot(x,x+3,'-r',label='-r')
plt.plot(x,x+4,'o',label='o')
plt.plot(x,x+5,'x',label='x')
plt.plot(x,x+6,'dr',label='dr')
#添加图例右下角 lower right 左上角 upper left 边框 透明度 阴影 边框宽度 
plt.legend(loc='lower right',fancybox=True,framealpha=1,shadow=True,borderpad=1)
plt.show()
```
