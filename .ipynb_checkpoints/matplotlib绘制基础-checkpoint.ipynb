```python
'''
 Matplotlib是一个Python的2D绘图层。通过matplotlib，开发者可以仅需要几行代码，便可以生成绘图，直方图，功率谱，条形图，错误图，散点图等。
'''
```

```python
'''
使用matplotlib绘制图形时，其中有两个最为常用的场景。一个是画点，一个是画线。
title()   设置图表名称
xlabel()   设置x轴名称
ylabel()    设置y轴名称
xticks(x,ticks,rotation)   设置x轴的刻度，rotation旋转角度
yticks()    设置y轴的刻度
plot()       绘制线性图表
show()       显示图表
legend()     显示图例
text(x,y,text)  显示每条数据的值 x，y值的位置
figure(name,figsize=(w,h),dpi=n)  设置图片大小
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([0,2],[1,4])
plt.show()
```

```python
'''
绘画折线
'''
x = [1,2,3,4,5]
squares = [1,4,9,16,25]
plt.plot(x,squares)
plt.show()
```

```python
x = [1,2,3,4,5]
squares = [1,4,9,16,25]
plt.plot(x,squares,linewidth=5)
plt.xlabel('datas',fontsize=14)
plt.ylabel('squares',fontsize=14)
plt.title('Number',fontsize=24)
plt.show()
```

```python
x = [1,2,3,4,5]
#y = x^2
y = [i*i for i in x]
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(200)
#y = x^2
y = [i*i for i in x]
plt.plot(x,y)
plt.savefig('result.jpg') #储存为图片
plt.show()
```
