```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,100) 
y=np.linspace(-10,10,100) #计算x和y的相交点a
x,y
```

```python
X,Y=np.meshgrid(x,y)
X.shape,Y.shape
```

```python
# 计算Z的坐标
Z=np.sqrt(X**2+Y**2)
Z.shape
```

```python
plt.contourf(X,Y,Z)
```

```python
plt.contour(X,Y,Z)
# 颜色越深表示值越小，中间的黑色表示 z=0.
plt.show()
```
